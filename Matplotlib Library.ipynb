# Plot


The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

```python
# line style 
#You can use the keyword argument linestyle, or shorter ls, to change the style of the plotted line:
plt.plot(x,y,color="green",linewidth=5 , linestyle = "dotted")
# x&y label 
plt.xlabel('Days')
plt.ylabel('Tem')
plt.title("Weather")
```

```python
# format string character & Markers 
plt.plot(x,y,"rD",ms=10,mec='black')
```

```python
#Matplotlib allows you to regulate the transparency of a graph plot using the alpha attribute.
plt.plot(x,y,color="red" ,alpha=0.2)
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
# create legend 
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("weather")
plt.plot(days,max_t , label="max")
plt.plot(days,min_t,label="min ")
plt.plot(days,avg_t , label = "avg")
# best loc of logend
plt.legend(loc="best" , shadow=True , fontsize="small")
plt.grid()
```

# Scatter 

The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
import matplotlib.pyplot as plt
import numpy as np
#The X-axis shows how old the car is.
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
#The Y-axis shows the speed of the car when it passes.
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

# Histogram

when we are working with a histogram you need only  single dimension array and the y_axis will contain the frequency 

```python

blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar , bins =3 , rwidth=0.96)
```

```python
# your bin parameter to specify the range explicitly 
# rwidth : relative width of abar compared to abin size 
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95 , color='r')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar, bins=[80,100,125,], rwidth=0.95, color='g' , histtype='step')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
# make side_by_side hist by change the first argument to list 
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
# make side_by_side hist by change the first argument to list 
plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'],orientation='horizontal')
plt.legend()
```

# PIE chart

With Pyplot, you can use the pie() function to draw pie charts:

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.axis("equal")
```

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True ,radius=.5)
plt.axis("equal")
plt.show()
```

```python
# use the autopct to supply will have percentage 
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5)
plt.show()
```

```python
# for take out these pies (start with home rent "counter clockwise ")
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0.1,0,0,0])
plt.show()
```

```python
# rotate for pie 
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0.1,0,0,0],counterclock=True, startangle=180)
plt.legend()
plt.show()

```

# Save matplotlib chart as afile 

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0.1,0,0,0],counterclock=True, startangle=180)
plt.savefig("piechart2.png" ,bbox_inches="tight" ,pad_inches=2,transparent=False)
```
